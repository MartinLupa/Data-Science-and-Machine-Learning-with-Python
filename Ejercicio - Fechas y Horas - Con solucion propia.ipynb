```python
import pandas as pd

file_path = "C:/Users/lupa_/OneDrive/Escritorio/Web Development/Data Science course/Seccion 4/UCI ADL Binary Dataset/OrdonezA_ADLs.txt"
```

```python
#Lectura de datos y creacion de columnas acorde al enunciado.
df = pd.read_csv(filepath_or_buffer = file_path,
                delimiter = r"\s+",
                header = 0)

df["extracted_start_date"] = df["Start"]
df["extracted_start_time"] = df["time"]
df["extracted_end_date"] = df["End"]
df["extracted_end_time"] = df["time.1"]


df
```

```python
#Lista conteniendo todas las columnas del DataFrame.
cols = df.columns.tolist()
cols
```

```python
#Para generar un string con las columnas deseadas y reasignarlas.
new_cols = [cols[4], cols[5], cols[6], cols[7], cols[8]] 
new_cols
```

```python
new_df = df[new_cols]
new_df = new_df.iloc[1:] #Sirve para obviar la primer fila, que no tiene informacion relevante.
new_df
```

```python
#Limpieza de filas cuya totalidad de valores sea NaN. Para este caso no existen, pero se realiza como practica.
new_df.dropna(axis = 0,
             how= "all",
             inplace = True) #El resultado de aplicar el metodo se vuelve a guardar en la variable new_df. 
new_df
```
